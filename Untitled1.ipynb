```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os

# Create a black image
img = np.zeros((511,511,3), np.uint8)


# Draw a diagonal blue line with thickness of 5 px
line = cv2.line(img,(0,0),(511,511),(255,0,0),5)
plt.imshow(line)
plt.show()
```

```python
img = np.zeros((511,511,3), np.uint8)
rectangle = cv2.rectangle(img,(400,200),(300,100),(0,255,0),3)
plt.imshow(rectangle)
plt.show()
```

```python
circle = cv2.circle(img,(447,63), 63, (0,0,255), -1)
plt.imshow(circle)
plt.show()
```

```python
ellipse = cv2.ellipse(img,(256,256),(100,50),0,0,180,255,-1)
plt.imshow(ellipse)
plt.show()
```

```python
# Height variable
from skimage.transform import rotate
for i in range (0, 20):
    square_size = 450
    img = np.zeros((square_size,square_size,3), np.uint8)
#     h = np.random.random_integers(square_size, size=(3,2))
    h = np.random.random_integers(300)
    w1 = np.random.random_integers(0,50)
    w2 = np.random.random_integers(200)
    pts = np.array([[200+w1/2,100+h],[100+w2/2,300],[400-w2/2,300],[300-w1/2,100+h]], np.int32)
    pts = pts.reshape((-1,1,2))
    polygon = cv2.polylines(img,[pts],True,(255,255,255),thickness=10)
    r = np.random.random_integers(180)
    polygon = rotate(polygon,r)

    print(w1)
    print(w2)
    plt.imshow(polygon*5)
    plt.show()
```
